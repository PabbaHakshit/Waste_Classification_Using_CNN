```python
import os
import cv2
import pandas as pd
from tqdm import tqdm
from glob import glob

x_data = []
y_data = []
for category in glob(train_path+'/*'):
    for file in tqdm(glob(category+'/*')):
        img_array = cv2.imread(file)
        img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)  # Fixed typo here
        x_data.append(img_array)
        y_data.append(os.path.basename(category))  # Use os.path.basename to get the category name

# Creating DataFrame
data = pd.DataFrame({'image': x_data, 'label': y_data})  # Fixed DataFrame syntax

print(data.head())  # Print the first few rows to check
```

```python
import os

# Explore training dataset
train_classes = os.listdir(train_path)
print("Classes in training set:", train_classes)

# Check the number of images per class
for cls in train_classes:
    print(f"Number of images in {cls}: {len(os.listdir(os.path.join(train_path, cls)))}")
```

```python
import matplotlib.pyplot as plt
import cv2
import random

# Visualize random images from the dataset
for cls in train_classes:
    class_path = os.path.join(train_path, cls)
    img_file = random.choice(os.listdir(class_path))
    img = cv2.imread(os.path.join(class_path, img_file))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(f"Class: {cls}")
    plt.show()
    break  # Remove to display more samples
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Preprocess images with resizing, normalization, and augmentation
data_gen = ImageDataGenerator(
    rescale=1./255,          # Normalize pixel values
    rotation_range=20,       # Augment with rotation
    width_shift_range=0.2,   # Horizontal shift
    height_shift_range=0.2,  # Vertical shift
    shear_range=0.2,         # Shear transformation
    zoom_range=0.2,          # Zoom augmentation
    horizontal_flip=True,    # Flip images horizontally
    validation_split=0.2     # Split for validation
)

train_data = data_gen.flow_from_directory(train_path, target_size=(128, 128), batch_size=32, class_mode='categorical', subset='training')
val_data = data_gen.flow_from_directory(train_path, target_size=(128, 128), batch_size=32, class_mode='categorical', subset='validation')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(train_classes), activation='softmax')  # Number of classes
])
model.summary()
```

```python
model.save('plastic_waste_model.h5')
print("Model saved!")
```
